```python
#import the libraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
import os
import requests
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
import json
```

```python
load_dotenv()
```

```python
my_gme = 0
headers = json.loads(open("keys.txt", 'r').read())
def hist_data(symbols, timeframe="15Min", limit=200, start="", end="", after="", until=""):
    df_data = {}
    bar_url = "https://data.alpaca.markets/v1" + f"/bars/{timeframe}"
    params = {"symbols":symbols,
             "limit":limit,
             "start":start,
             "end":end,
             "after":after,
             "until":until}
    r = requests.get(bar_url, headers=headers, params=params)
    json_dump = r.json()
    for symbol in json_dump:
        temp_df = pd.DataFrame(json_dump[symbol])
        temp_df.rename({"t":"time", "o":"open", "h":"high","l":"low","c":"close","v":"volume"}, axis=1, inplace=True)
        temp_df['time'] = pd.to_datetime(temp_df['time'], unit="s")
        temp_df.set_index("time",inplace=True)
        temp_df.index = temp_df.index.tz_localize("UTC").tz_convert("America/Indiana/Petersburg")
        df_data[symbol] = temp_df
    return df_data

gme_df = hist_data('GME,AAPL')['GME']
gme_df
```

```python
# Set index as datetime object and drop extraneous columns
# Drop NAs and calculate daily percent return
gme_df['daily_return'] = gme_df['close'].dropna().pct_change()
gme_df
```

```python
# Set short and long windows
short_window = 1
long_window = 10

# Construct a `Fast` and `Slow` Exponential Moving Average from short and long windows, respectively
gme_df['fast_close'] = gme_df['close'].ewm(halflife=short_window).mean()
gme_df['slow_close'] = gme_df['close'].ewm(halflife=long_window).mean()

# Construct a crossover trading signal
gme_df['crossover_long'] = np.where(gme_df['fast_close'] > gme_df['slow_close'], 1.0, 0.0)
gme_df['crossover_short'] = np.where(gme_df['fast_close'] < gme_df['slow_close'], -1.0, 0.0)
gme_df['crossover_signal'] = gme_df['crossover_long'] + gme_df['crossover_short']

gme_df.head()
```

```python
# Plot the EMA of BTC/USD closing prices
gme_df[['close', 'fast_close', 'slow_close']].plot(figsize=(20,10))
```

```python
### Construct Exponential Moving Average of Daily Return Volatility (Trading Signal, Feature Two)
# Set short and long volatility windows
short_vol_window = 1
long_vol_window = 10

# Construct a `Fast` and `Slow` Exponential Moving Average from short and long windows, respectively
gme_df['fast_vol'] = gme_df['daily_return'].ewm(halflife=short_vol_window).std()
gme_df['slow_vol'] = gme_df['daily_return'].ewm(halflife=long_vol_window).std()

# Construct a crossover trading signal
gme_df['vol_trend_long'] = np.where(gme_df['fast_vol'] < gme_df['slow_vol'], 1.0, 0.0)
gme_df['vol_trend_short'] = np.where(gme_df['fast_vol'] > gme_df['slow_vol'], -1.0, 0.0) 
gme_df['vol_trend_signal'] = gme_df['vol_trend_long'] + gme_df['vol_trend_short']

gme_df
```

```python
### Plot Exponential Moving Average of Daily Return Volatility
```

```python
# Plot the EMA of BTC/USD daily return volatility
gme_df[['fast_vol', 'slow_vol']].plot(figsize=(40,20))
```

```python
### Construct Bollinger Band (Trading Signal, Feature Three)
```

```python
# Set bollinger band window
bollinger_window = 20

# Calculate rolling mean and standard deviation
gme_df['bollinger_mid_band'] = gme_df['close'].rolling(window=bollinger_window).mean()
gme_df['bollinger_std'] = gme_df['close'].rolling(window=20).std()

# Calculate upper and lowers bands of bollinger band
gme_df['bollinger_upper_band']  = gme_df['bollinger_mid_band'] + (gme_df['bollinger_std'] * 1)
gme_df['bollinger_lower_band']  = gme_df['bollinger_mid_band'] - (gme_df['bollinger_std'] * 1)

# Calculate bollinger band trading signal
gme_df['bollinger_long'] = np.where(gme_df['close'] < gme_df['bollinger_lower_band'], 1.0, 0.0)
gme_df['bollinger_short'] = np.where(gme_df['close'] > gme_df['bollinger_upper_band'], -1.0, 0.0)
gme_df['bollinger_signal'] = gme_df['bollinger_long'] + gme_df['bollinger_short']

gme_df
```

```python
### Plot Bollinger Bands
```

```python
# Plot the Bollinger Bands for BTC/USD closing prices
gme_df[['close','bollinger_mid_band','bollinger_upper_band','bollinger_lower_band']].plot(figsize=(20,10))
```
