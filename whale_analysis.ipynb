# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
csv_path = Path("Resources/whale_returns.csv")
whale_df = pd.read_csv(csv_path, index_col="Date",parse_dates=True, infer_datetime_format=True)
```

```python
##Sort in ascending order and 

whale_df.sort_index(ascending = True, inplace = True)
whale_df.head()
```

```python
# Count nulls
whale_df.isnull().sum()
```

```python
# Drop nulls
whale_df.dropna(inplace=True)
```

```python
whale_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_df_csvpath = Path("Resources/algo_returns.csv")
algo_df = pd.read_csv(algo_df_csvpath, index_col="Date",parse_dates=True,infer_datetime_format = True)
algo_df.sort_index(ascending=True,inplace = True)
```

```python
# Count nulls
algo_df.isnull().sum()
```

```python
# Drop nulls
algo_df.dropna(inplace=True)
algo_df
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_csvpath = Path("Resources/sp500_history.csv")
sp500_df = pd.read_csv(sp500_csvpath,index_col="Date",parse_dates=True,infer_datetime_format = True)
sp500_df.sort_index(ascending=True, inplace = True)
sp500_df.head()
```

```python
# Check Data Types
sp500_df.dtypes
```

```python
# Fix Data Types
sp500_df["Close"]= sp500_df["Close"].str.replace('$','').astype("float")
sp500_df.sort_index(inplace = True)
sp500_df

```

```python
# Calculate Daily Returns
sp500_daily_returns = sp500_df.pct_change()
```

```python
# Drop nulls
sp500_daily_returns.dropna(inplace = True)
sp500_daily_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp500_daily_returns.rename(columns = {"Close":"S&P500 Daily Returns"}, inplace = True)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
joined_df_daily = pd.concat([whale_df,algo_df,sp500_daily_returns], axis = "columns",join = "inner")
joined_df_daily
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = df_daily_std*np.sqrt(252)
annualized_std
```

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
sp500_rolling_21 = joined_df_daily[["S&P500 Daily Returns"]].rolling(window = 21).std()
```

```python
# Plot the rolling standard deviation
sp500_rolling_21.plot(figsize = (20,10))
```

### Calculate and plot the correlation

```python
# Calculate the correlation
jpr_correlation = joined_df_daily.corr()
jpr_correlation
```

```python
# Display de correlation matrix
sns.heatmap(jpr_correlation, vmin=-1, vmax=1);
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
tiger_sp500_covariance = joined_df_daily['TIGER GLOBAL MANAGEMENT LLC'].cov(joined_df_daily['S&P500 Daily Returns'])
tiger_sp500_covariance
```

```python
# Calculate variance of S&P 500
sp500_var = joined_df_daily['S&P500 Daily Returns'].var()
sp500_var
```

```python
# Computing beta
tiger_beta = tiger_sp500_covariance / sp500_var
tiger_beta
```

```python
# Plot beta trend
sns.lmplot(x='S&P500 Daily Returns', y='TIGER GLOBAL MANAGEMENT LLC', data=joined_df_daily, aspect=1.5, fit_reg=True);
```
